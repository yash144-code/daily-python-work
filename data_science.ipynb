### More Complex `for` loop Example

This example uses a `for` loop to process a list of student dictionaries. For each student, it checks their `score`. If the score is 60 or higher, it prints a 'Pass' message; otherwise, it prints 'Fail'.

```python
students = [
    {"name": "Alice", "score": 85},
    {"name": "Bob", "score": 55},
    {"name": "Charlie", "score": 70},
    {"name": "Diana", "score": 40}
]

print("--- Student Performance ---")
for student in students:
  name = student["name"]
  score = student["score"]

  if score >= 60:
    print(f"{name} passed with a score of {score}.")
  else:
    print(f"{name} failed with a score of {score}.")
print("---------------------------")
```

```python
import numpy as np
import matplotlib.pyplot as plt
hours = np.array([1, 2, 3, 4, 5])
marks = np.array([35, 45, 55, 65, 75])
m, c = np.polyfit(hours, marks, 1)
plt.scatter(hours, marks, label="Data points")

# trend line values
y_line = m * hours + c

plt.plot(hours, y_line, color="red", label="Trend line")

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks with Trend Line")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

ad_spend = np.array([5, 8, 10, 4, 7, 12, 15, 9, 6, 11, 10, 16, 3, 14, 8, 11, 15, 6, 13, 17])  # in lakhs
sales = np.array([20, 30, 38, 18, 28, 45, 55, 34, 24, 42, 40, 60, 15, 52, 32, 44, 56, 25, 48, 65])  # in lakhs

m,c = np.polyfit(ad_spend,sales,1)

plt.scatter(ad_spend,sales,marker = "^",label = "data points")

y_line = m * ad_spend + c

plt.plot(ad_spend,y_line,color = "r",linestyle = "-." ,label = "Trend Line")

plt.xlabel('ad_spend')
plt.ylabel('sales')
plt.title('sales by ads')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("fruits1.csv")
print(df.head())
print("----------------------------------------")
m = df['Orange'].mean()
app = df['Apples'].mean()
bana = df['Banana'].mean()
fill_values = {'Orange': m, 'Apples': app, 'Banana': bana}
df[['Orange', 'Apples', 'Banana']] = df[['Orange', 'Apples', 'Banana']].fillna(fill_values)

print(df)
```
