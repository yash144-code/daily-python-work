### Question: Check if a string is a palindrome.

Write a Python program that asks the user to enter a string. Based on the input, determine if the string is a palindrome (reads the same forwards and backwards, ignoring case and spaces), and print an appropriate message.

```python
input_string = input("Enter a string: ")

# Clean the string: convert to lowercase and remove spaces
cleaned_string = input_string.replace(" ", "").lower()

# Check if the cleaned string is equal to its reverse
if cleaned_string == cleaned_string[::-1]:
    print(f'"{input_string}" is a palindrome.')
else:
    print(f'"{input_string}" is not a palindrome,')
```

3.	Print only expenses above 300


```python
expenses = [200, 450, 120, 560, 300, 90, 1000]
#1  find total expense
total_expenses= 0
for e in expenses:
  total_expenses+= e
print("The total expenses:",total_expenses)
print()

#2. Find how many days expense > 300
count = 0
for e in expenses:
  if e>300:
    count +=1
print("no.of days e>300 is: ",count)
print()


#3. Print only expenses above 300

for e in expenses:
  if e > 300:
    print(e)
```

Problem 1: Smart Expense Analyzer  

You are given daily expenses of a month:

expenses = [120, 340, 560, 80, 900, 450, 230, 1500, 60, 700]


Do the following:

Find total expense

Find average expense

Count high-spending days (â‰¥ 500)

Print days (index) where expense is > 500

Find the maximum expense

```python
expenses = [120, 340, 560, 80, 900, 450, 230, 1500, 60, 700]

total = 0
high_count = 0
max_expense = expenses[0]

for i in range(len(expenses)):
    total += expenses[i]

    if expenses[i] >= 500:
        high_count += 1
        print("High spending on day", i, ":", expenses[i])

    if expenses[i] > max_expense:
        max_expense = expenses[i]

average = total / len(expenses)

print("\nTotal expense:", total)
print("Average expense:", average)
print("High spending days count:", high_count)
print("Maximum expense:", max_expense)
```
