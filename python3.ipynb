### Practice Question: Determine the grade of a student

Write a Python program that declares an integer variable called `score` and assigns it a value between 0 and 100. Use `if-elif-elif-else` statements to determine and print the corresponding grade

```python
marks = int(input("enter marks:"))
if marks >=90:
  print("Grade A")
elif marks>=80:
  print("Grade B")
elif marks>=70:
  print("grade C")
elif marks>=60:
  print("grade D")
else:
  print("grade F")
```

### Practice Question: Determine the type of triangle

Write a Python program that declares three integer variables, `side1`, `side2`, and `side3`, representing the lengths of the sides of a triangle. Assign them appropriate values. Use `if-elif-else` statements to determine and print the type of triangle based on the following criteria:

*   If all three sides are equal: "Equilateral Triangle"
*   If any two sides are equal: "Isosceles Triangle"
*   If all three sides are different: "Scalene Triangle"

Test your code with different values for the sides (e.g., (5, 5, 5), (4, 4, 6), (3, 4, 5)). Remember that for a valid triangle, the sum of any two sides must be greater than the third side (you don't necessarily need to implement this validation unless you want an extra challenge).

```python
side1 = int(input("enter lenght:"))
side2 = int(input("enter lenght:"))
side3 = int(input("enter lenght:"))
if side1==side2==side3:
  print("Equilateral Triangle")
elif side1==side2 or side2==side3 or side3==side1:
  print( "Isosceles Triangle")
else:
  print("Scalene Triangle")
```
